```python
from functions.get_function_setting import get_function_setting
from algorithms import genetic_algorithm
from matplotlib import pyplot as plt
```

```python
# number_iteration: 500
for number in range(1, 7+1, 1):
    function_setting = get_function_setting(number)

    # parameter setting
    number_iteration = 500
    number_bit = 16
    number_population = 50
    cross_rate = 0.9
    mutation_rate = 1 / (number_bit * function_setting["dimension"])
    boundaries = [function_setting["boundary"]] * function_setting["dimension"]

    best_solution, best_fitness, best_fitness_set = \
        genetic_algorithm(
            function_setting["function"],
            boundaries,
            number_bit,
            number_iteration,
            number_population,
            cross_rate,
            mutation_rate
        )

    print(f"Best fitness of function {number} is: {best_fitness}")
    # print(f"Best solution is: {best_solution}")
    plt.plot(best_fitness_set)
    plt.show()
```
